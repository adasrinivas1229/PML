**Converting 'airline_sentiment' to neumeric** 

Machine learning models work best when inputs are numerical. we will convert all the chosen columns to their needed numerical formats.

Transform **airline_sentiment** column to numerical category:

Calling the factorize method returns an array of numeric categories and an index of the categories. In this case, index 0 is for **positive** and index 1 is **negative** sentiment respectively.



```python
sentiment_label = tweet_df.airline_sentiment.factorize()
sentiment_label
```

```python
# alternate method of label encoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
tweet_df['airline_sentiment']= label_encoder.fit_transform(tweet_df['airline_sentiment']) 
np.unique(tweet_df['airline_sentiment'])
```

**Data Preprocessing**

From the following code:

we get the actual texts from the data frame.
Initialize the tokenizer with a 5000 word limit. This is the number of words we would like to encode.
we call fit_on_texts to create associations of words and numbers.
calling text_to_sequence replaces the words in a sentence with their respective associated numbers. This transforms each sentence into sequences of numbers. The sentences or tweets have different number of words, therefore, the length of the sequence of numbers will be different.
Our model requires inputs to have equal lengths, so we will have to pad the sequence to have the chosen length of inputs. This is done by calling the pad_sequence method with a length of 200.
All input sequences will have a length of 200.

```python
#imporiting necessary libraries
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tweet = tweet_df.text.values          #Getting values of tweet_df.text to tweet
tokenizer = Tokenizer(num_words=5000) #creating  Tokenizer object with 
                                      #maximum 5000 word Limit
#converting text as number of tokens
tokenizer.fit_on_texts(tweet)         
#Set vocab_size with the lengths of no. of words
vocab_size = len(tokenizer.word_index) + 1
#Converting Text to integer Sequences 
encoded_docs = tokenizer.texts_to_sequences(tweet)
#pad sequences to the same length
padded_sequence = pad_sequences(encoded_docs, maxlen=200)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
tweet = tweet_df.text.values
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(tweet)
vocab_size = len(tokenizer.word_index) + 1
encoded_docs = tokenizer.texts_to_sequences(tweet)
padded_sequence = pad_sequences(encoded_docs, maxlen=200)
```

# **Build the Model** 

The model consists of an embedding layer, LSTM layer and a Dense layer which is a fully connected neural network with sigmoid as the activation function. Dropouts are added in-between layers and also on the LSTM layer to avoid overfitting. 


```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import LSTM,Dense, Dropout
from tensorflow.keras.layers import SpatialDropout1D
from tensorflow.keras.layers import Embedding
embedding_vector_length = 32
#Sequential groups a linear stack of layers into a tf.keras.Model.
model = Sequential()     
# Embedding turns positive integers (indexes) into dense vectors of fixed size.
 # number of parameters are vocab (13234) * embedding_vector_length(32) = 423488
model.add(Embedding(vocab_size, embedding_vector_length,   
                                     input_length=200) )    
#Adding Spatial Dropout layer with 1/4th dropout ratio to drop 1D features                                          
model.add(SpatialDropout1D(0.25))
#Adding LSTM Layer  with dimensionality of the output shape = 50 
#dropout ratio for input=50% dropout ratio for recurrent state=50%
model.add(LSTM(50, dropout=0.5, recurrent_dropout=0.5))
#adding dropout layer with 20% dropout ratio
model.add(Dropout(0.2))
#add a dense layer with output shape '1' , use activation function sigmoid
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam', 
                           metrics=['accuracy'])
```

```python
print(model.summary())
```

## **Train the Model**

The model is trained for 5 epochs which attains a validation accuracy of ~92%.

```python
history = model.fit(padded_sequence,sentiment_label[0],
                  validation_split=0.2, epochs=5, batch_size=32)
```

## **Test the Model**

The model is tested with a sample text to see how it predicts sentiment and we can see that it predicted the right sentiment for the sentence.

```python
test_word ="bad Experience"
tw = tokenizer.texts_to_sequences([test_word])
tw = np.array(pad_sequences(tw,maxlen=200))
#p=model.predict_classes(tw)
prediction = int(model.predict(tw).round().item())
sentiment_label[1][prediction]
```
